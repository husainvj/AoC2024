```python

def is_valid_position(row, col):
    """Check if a position is within grid bounds."""
    return 0 <= row < len(lines) and 0 <= col < len(lines[0])
```

```python
def check_sequence_from(row, col, direction, sequence):
    """
    Check if the sequence appears starting at (row, col) in the given direction.
    """
    for k in range(len(sequence)):
        new_row = row + k * direction[0]
        new_col = col + k * direction[1]
        if not is_valid_position(new_row, new_col) or lines[new_row][new_col] != sequence[k]:
            return False
    return True
```

```python
new_directions = [
    (-1, -1), (-1, 1),  # Top-left, Top-right
    (1, -1),  (1, 1)     # Bottom-left, Bottom-right
]
```

```python
def check_cross_sequence_from(row, col, direction, sequence):
    """
    Check if the sequence appears starting at (row, col) in the given direction.
    """
    for k in range(len(sequence)):
        new_row = row + k * direction[0]
        new_col = col + k * direction[1]
        if not is_valid_position(new_row, new_col) or lines[new_row][new_col] != sequence[k]:
            return False
    return lines[row + direction[0]][col + direction[1]] == 'A'             # here we return the cordiantes of A rather tha just True 
```

```python
found = False
new_coordinates = []
new_count = []
for i in range(len(test_lines)):
    for j in range(len(test_lines[0])):
        if test_lines[i][j] == "M":  # Check if the first letter matches
            for direction in new_directions:  # Check in 4 diagonal directions
                if check_cross_sequence_from(i, j, direction, "MAS"):
                    
                    print(f"Sequence '{"MAS"}' found starting at ({i}, {j}) in direction {direction}")
                    found = True
                    new_count.append(1)
                    new_coordinates.append((i,j))
                    

if not found:
    test_lines[i][j].replace( test_lines[i][j], ".")
```
