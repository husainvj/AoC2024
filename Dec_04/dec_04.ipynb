```python
def check_sequence_from(row, col, direction, sequence):
    """
    Check if the sequence appears starting at (row, col) in the given direction.
    """
    for k in range(len(sequence)):
        new_row = row + k * direction[0]
        new_col = col + k * direction[1]
        if not is_valid_position(new_row, new_col) or lines[new_row][new_col] != sequence[k]:
            return False
    return True
```
