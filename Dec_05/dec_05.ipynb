In this part we used the previous if condition to just find out the incorrect ones and make them a list. We keep doing this and eventually end up with all pairs correctly ordered. Now we change the corrected pairs into an ordered list by sorting them by how many times each of them appears in the second place of the pairs. This gives us an ordered list of numbers which we can verify by running the /make_pairs and if all() function to check. However, we are missing the first number in the oredered list somehow, so we substracted 1 from the index to get the middle number.     

```python
incorrect_list = []
for i in range(len(pg_nos_list)):
    temp_pattern = [[pg_nos_list[i][x], pg_nos_list[i][y]] for x in range(len(pg_nos_list[i])) for y in range(x + 1, len(pg_nos_list[i]))]
    if not all(pair in pattern_list for pair in temp_pattern):
        incorrect_list.append(pg_nos_list[i])
```

```python
def make_pairs(lst):
    return [[lst[x], lst[y]] for x in range(len(lst)) for y in range(x + 1, len(lst))]
```

```python
corrected_list = []
#new_mids = []
for g in range(len(incorrect_list)):
    for n in range(len(incorrect_list[g])):
        temp_list=make_pairs(incorrect_list[g])
        for k in range(len(temp_list)):
            if temp_list[k] not in pattern_list:
                temp_list[k] = temp_list[k][::-1]
        temp_list
    second_elements = [pair[1] for pair in temp_list]
    counts = Counter(second_elements)
    corrected_list.append([value for value, count in sorted(counts.items(), key=lambda x: x[1])])
                                  
```
